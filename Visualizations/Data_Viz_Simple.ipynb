```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('posts_Year_2022_01_updated.csv')

df.plot(x='name', y='totalCount', kind='bar')

plt.title('Total Count by Name')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.plot(kind='scatter', x='votesCount', y='reviewsRating')

plt.title('Reviews Rating vs Votes Count')
plt.show()
```

```python
topics = df['topics'].str.split(',', expand=True).stack().value_counts()

topics.plot(kind='pie')

plt.title('Topics')
plt.show()
```

```python
df['createdAt'] = pd.to_datetime(df['createdAt'])
df.set_index('createdAt').resample('D').count()['id'].plot()

plt.title('Posts Created Over Time')
plt.show()
```
