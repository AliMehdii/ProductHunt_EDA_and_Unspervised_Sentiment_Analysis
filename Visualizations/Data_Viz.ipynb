```python
import numpy as np
import pandas as pd
from plotly.express import sunburst
import matplotlib.pyplot as plt
import seaborn as sns

# df = pd.read_csv('posts_After_2022-01-01_new copy.csv')
df = pd.read_csv('posts_Year_2022_01_updated.csv')

fig = sunburst(df, path=['topics'], values='totalCount')
fig.show()
```

```python
sns.set()
sns.pairplot(df, diag_kind="kde")
plt.show()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('posts_Year_2022_01_updated.csv')

fig = px.treemap(df, path=['topics'], values='totalCount')
fig.show()
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('posts_Year_2022_01_updated.csv')

fig = px.parallel_categories(df, dimensions=['topics', 'votesCount', 'commentsCount'])
fig.show()
```

```python
import pandas as pd
import plotly.express as px 

df = pd.read_csv('posts_Year_2022_01_updated.csv')

fig = px.scatter_3d(df, x='votesCount', y='commentsCount', z='topics',
                   color='topics')
fig.show()
```

```python
from wordcloud import WordCloud
import pandas as pd

df = pd.read_csv('posts_Year_2022_01.csv')
text = " ".join(df['topics'])

wordcloud = WordCloud().generate(text)

import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import pandas as pd
from plotly.express import pie

df = pd.read_csv('posts_Year_2022_01_updated.csv')

fig = pie(df, values='totalCount', names='topics')
fig.update_traces(textposition='inside')
fig.show()
```
